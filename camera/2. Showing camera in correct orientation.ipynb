## Let's do some exercise

### Import some libraries

````python
import cv2
from IPython.display import Image, display
````

### Load an image and display it via cv2

````python
# Load the image
image = cv2.imread('resource/dgirl_300.jpg')

# Display the image
display(Image(data=cv2.imencode('.jpg', image)[1].tobytes()))
````

### How to flip the image vertically?

We flip the image by the X axis (Code 0)

````python
# Load the image
image = cv2.imread('resource/dgirl_300.jpg')
image = cv2.flip(image,0)

# Display the image
display(Image(data=cv2.imencode('.jpg', image)[1].tobytes()))
````

### How to flip the image horizontally?

We flip the image by the Y axis (Code 1)

````python
# Load the image
image = cv2.imread('resource/dgirl_300.jpg')
image = cv2.flip(image,1)

# Display the image
display(Image(data=cv2.imencode('.jpg', image)[1].tobytes()))
````

### How to flip the image both vertically and horizontally?

We flip the image by the X and Y axis (Code -1)

Replace the `???` in the code below with the correct code to flip both axes

````python
# Load the image
image = cv2.imread('resource/dgirl_300.jpg')
image = cv2.flip(image, -1)

# Display the image
display(Image(data=cv2.imencode('.jpg', image)[1].tobytes()))
````

## Fix webcam orientation

Fix the program below to make the camera showing the correct orientation. You just need to add one line to the program below.

`frame = cv2.flip(frame, -1)`

````python
cap = cv2.VideoCapture(0)
    
ret, frame = cap.read()

if ret:
    _, frame = cv2.imencode('.jpeg', frame)
    display(Image(data=frame.tobytes()))
else:
    print("failed")

# Release the webcam
cap.release()
````

### Answer

````python
cap = cv2.VideoCapture(0)
    
ret, frame = cap.read()

if ret:
    frame = cv2.flip(frame, 0)
    _, frame = cv2.imencode('.jpeg', frame)
    display(Image(data=frame.tobytes()))
else:
    print("failed")

# Release the webcam
cap.release()

````
