```python
import matplotlib.pyplot as plt
import numpy as np

# Define three points
points = np.array([[5, 2], [3, 4], [5, 6]])

# Plot the points
plt.plot(points[:, 0], points[:, 1], 'bo')

# Define the lines passing through the points
line1 = points[1] - points[0]
line2 = points[2] - points[1]

# Calculate the angle between the lines
angle = np.arccos(np.dot(line1, line2) / (np.linalg.norm(line1) * np.linalg.norm(line2)))
angle_deg = np.degrees(angle)

# Annotate the angle between the lines
plt.text(points[1][0], points[1][1], f'{angle_deg:.2f}°', fontsize=12, ha='right')

# Plot the lines passing through the points
plt.plot([points[0][0], points[1][0]], [points[0][1], points[1][1]], 'r-')
plt.plot([points[1][0], points[2][0]], [points[1][1], points[2][1]], 'r-')

# Set plot limits
plt.xlim(0, 7)
plt.ylim(0, 8)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Angle between two lines')

# Show the plot
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Arc

# Define three points
points = np.array([[4, 2], [3, 4], [5, 6]])

# Plot the points
plt.plot(points[:, 0], points[:, 1], 'bo')

# Define the lines passing through the points
line1 = points[1] - points[0]
line2 = points[2] - points[1]

# Calculate the angle between the lines
angle = np.arccos(np.dot(line1, line2) / (np.linalg.norm(line1) * np.linalg.norm(line2)))
angle_deg = np.degrees(angle)

# Annotate the angle between the lines
plt.text(points[1][0], points[1][1], f'{angle_deg:.2f}°', fontsize=12, ha='right')

# Plot the lines passing through the points
plt.plot([points[0][0], points[1][0]], [points[0][1], points[1][1]], 'r-')
plt.plot([points[1][0], points[2][0]], [points[1][1], points[2][1]], 'r-')

# Calculate the position and angle for the arc
arc_center = points[1]
arc_radius = 1.5
start_angle = np.arctan2(points[0][1] - arc_center[1], points[0][0] - arc_center[0])
start_angle_deg = np.degrees(start_angle)
if start_angle_deg < 0:
    start_angle_deg += 360

# Draw an arc to indicate the angle
arc = Arc(arc_center, arc_radius*2, arc_radius*2, angle=start_angle_deg, theta1=0, theta2=angle_deg, color='k')
plt.gca().add_patch(arc)

# Label the angle as 'theta'
plt.text(arc_center[0] + 0.5, arc_center[1] + 0.5, r'$\theta$', fontsize=12)

# Set plot limits
plt.xlim(0, 7)
plt.ylim(0, 8)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Angle between two lines')

# Show the plot
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Arc, Circle

# Define three points
points = np.array([[4, 2], [3, 4], [5, 6]])

# Plot the points
plt.plot(points[:, 0], points[:, 1], 'bo')

# Define the vectors passing through the points
vector1 = points[1] - points[0]
vector2 = points[2] - points[1]

# Calculate the angle between the vectors
cos_angle = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
angle_rad = np.arccos(np.clip(cos_angle, -1.0, 1.0))
angle_deg = np.degrees(angle_rad)

# Annotate the angle between the lines
plt.text(points[1][0], points[1][1], f'{angle_deg:.2f}°', fontsize=12, ha='right')

# Plot the lines passing through the points
plt.plot([points[0][0], points[1][0]], [points[0][1], points[1][1]], 'r-')
plt.plot([points[1][0], points[2][0]], [points[1][1], points[2][1]], 'r-')

# Calculate the position and angle for the arc
arc_center = points[1]
arc_radius = 1.5
start_angle = np.arctan2(points[0][1] - arc_center[1], points[0][0] - arc_center[0])
start_angle_deg = np.degrees(start_angle)
if start_angle_deg < 0:
    start_angle_deg += 360

# Draw an arc to indicate the angle
arc = Arc(arc_center, arc_radius*2, arc_radius*2, angle=start_angle_deg, theta1=0, theta2=angle_deg, color='k')
plt.gca().add_patch(arc)

# Label the angle as 'theta'
plt.text(arc_center[0] + 0.5, arc_center[1] + 0.5, r'$\theta$', fontsize=12)

# Define the center of the circle
circle_center = (4, 8)
circle_radius = 0.5

# Draw a circle
circle = Circle(circle_center, circle_radius, edgecolor='b', facecolor='none')

# Add the circle to the plot
plt.gca().add_patch(circle)

# Set plot limits
plt.xlim(0, 10)
plt.ylim(0, 10)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Angle between two lines with a circle at (8, 4)')

# Show the plot
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
